```python
#
# 
# 
#
# partial release history:
# 2022/1/18 m.maruyama   Created
#
import csv, os
import pandas as pd
import glob
import numpy as np
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import datetime
from datetime import date, timedelta

#CSVdf
# CSV
files = sorted(glob.glob('csv/*.csv'))
# 
file_number = len(files)
# CSV
csv_list = []
for file in files:
    csv_list.append(pd.read_csv(file,encoding='shift_jis',skiprows=[1]))
# CSV
covid_data_org = pd.concat(csv_list)

covid_data = covid_data_org.copy()
#
covid_data.drop([
    "", 
    "", 
    "", 
    "_", 
    "_", 
    "_", 
    "_", 
    "_", 
    "_", 
    "_"], axis=1,inplace=True)

#
covid_data = covid_data_org.groupby(['_', '_'], as_index=False).count()

# 
#covid_data.to_csv('covid_data_out.csv')

# df
# 
columns1 =[ "100",
            "10",
            "10",
            "20",
            "30",
            "40",
            "50",
            "60",
            "70",
            "80",
            "90"]

# 2021/10/1
#d1 = date(2020,2,26)
d1 = date(2021,10,1)
d2 = datetime.date.today()

# X
xDate = []
for i in range((d2 - d1).days + 1):
    date = d1 + timedelta(i)
    xDate.append(date)
    
# Y
yData = []

# 
for col, old in enumerate(columns1):
    covid_df1 = covid_data[covid_data['_'] == old]
    yDataOld = []

    # 
    for row, date1 in enumerate(xDate):
        strdate = date1.strftime('%Y/%m/%d')
        strdate = strdate.replace('/0', '/')
        covid_df2 = covid_df1[covid_df1['_'] == strdate]
        if len(covid_df2) :
            yDataOld.append(covid_df2.iat[0, 2])
        else :
            yDataOld.append(0)
    # Y
    yData.append(yDataOld)
        
# 
plt.figure(figsize=(12, 6))
plt.stackplot(
        xDate, 
        yData,
        labels = columns1,
        colors = None
)

# 
dt_now = datetime.datetime.now()
plt.title("" + dt_now.strftime('%Y-%m-%d %H:%M:%S') + " ", fontname="MS Gothic")

# 
plt.xlabel("", fontname="MS Gothic")
plt.ylabel("", fontname="MS Gothic")

# 
plt.yticks(np.arange(0, 500, step=50))

# 
plt.legend(	bbox_to_anchor=(1.05, 1), 
		loc='upper left', 
		borderaxespad=0, 
		fontsize=12, 
		ncol=1, 
		prop={"family":"MS Gothic"})

# 
strfname = 'png/' + d2.strftime('%Y%m%d') + '_covidGifu_oldStack.png'
plt.savefig(strfname, bbox_inches='tight')

plt.show()
plt.close('all')
```
